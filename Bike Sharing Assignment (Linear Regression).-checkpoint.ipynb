# <i> Processing Steps </i>

### Creating Dummies 

```python
# Creating the dummy variables for the categorical variables having more than 2 unique values.
# The dummy variables will be (n-1)no. of variables for  the specified columns .
```

```python
s=pd.get_dummies(day["season"],drop_first=True)
m=pd.get_dummies(day["mnth"],drop_first=True)
w=pd.get_dummies(day["weekday"],drop_first=True)
weath=pd.get_dummies(day["weathersit"],drop_first=True) 
```

```python
 # As these variables are converted to dummies so droping these variables.
day.drop(["season","mnth","weekday","weathersit"],axis=1,inplace=True)
```

```python
# Concatenating the dummies in the data frame.
day=pd.concat([day,s,m,w,weath],axis=1)
```

```python
day.head()             #Inspecting the dataframes
```

```python
#As the count ("cnt") column is a sum of both the variables as they are dependent to each other so droping these variables.
day.drop(["registered","casual"],axis=1,inplace=True)
```

# <i> Importing the libraries for regression analysis</i>

```python
import statsmodels
import statsmodels.api as sm
import sklearn
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

### Spliting the models into train and test  (70- 30)  set

```python
np.random.seed(0)
day_train,day_test=train_test_split(day,train_size=0.7,random_state=100)
```

## Rescaling the features using MinMaxscaling 

```python
#Scaling the train datasets
Scaler=MinMaxScaler()
day_train[num_col]=Scaler.fit_transform(day_train[num_col])
```

## Dividing into X and y sets for the model building

```python
y_train=day_train.pop('cnt')
X_train=day_train
```

# <i>Building model</i>

```python
#RFE feature of sklearn will help to build the linear regression
```

```python
lm=LinearRegression()
lm.fit(X_train,y_train)
rfe=RFE(lm,16)               #Running the RFE with 16 variables as output
rfe=rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
new_col=X_train.columns[rfe.support_]    #New Columns after RFE feature estimation
```

```python
X_train_new=X_train[new_col]             #giving a new variable name 
```

```python
X_train_new.columns                      #inspecting the columns
```

```python
X_train_new                            # head of the train data set
```

### Model1

```python
X_train_sm=sm.add_constant(X_train_new)   # As we have to add the constant because its take the intercept from the origin by default.
lr=sm.OLS(y_train,X_train_sm)         
lr_model_1=lr.fit()                 #fitting the model for the prediction
```

```python
lr_model_1.summary()                # Summary of the model
```

```python
# Calculating the Variance Inflation Factor  
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

The p-value greater than 0.05 and the VIF grater than 5 is insignificant for the variable . It also to be noticed that the p-value is considered as the first than we consider the VIF.
As we have observed from the above that the p-value for the "Feb" is greater than 0.05 which not considered as the good factor so we will drop the particular column and will re model it again .

```python
X_train_new=X_train_new.drop("Feb",axis=1) # Droping the variable "Feb"
```

## Model2

```python
X_train_sm=sm.add_constant(X_train_new)   # Again repeating the same 
lr=sm.OLS(y_train,X_train_sm)
lr_model_2=lr.fit()
lr_model_2.summary()
```

```python
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

Here we observed that the p-value for the "Dec" is close to be 0.05 which is not considered as good factor for the prediction so we delete the variable.

```python
X_train_new=X_train_new.drop("Dec",axis=1)   # Droping the "dec" variable.
```

### Model 3 

```python
X_train_sm=sm.add_constant(X_train_new)
lr=sm.OLS(y_train,X_train_sm)
lr_model_3=lr.fit()
lr_model_3.summary()
```

```python
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X_train_new=X_train_new.drop("Nov",axis=1)
```

Here we get the p-value for the "Nov" column as greater than 0.05 so we will drop the variable "Nov"

### Model 4

```python
X_train_sm=sm.add_constant(X_train_new)
lr=sm.OLS(y_train,X_train_sm)
lr_model_4=lr.fit()
lr_model_4.summary()
```

```python
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

Again we get the high p-value for the "Jan" column as it is greater than 0.05. So we will drop the variable "Jan".

```python
X_train_new=X_train_new.drop("Jan",axis=1)
```

### Model 5 

```python
X_train_sm=sm.add_constant(X_train_new)
lr=sm.OLS(y_train,X_train_sm)
lr_model_5=lr.fit()
lr_model_5.summary()
```

```python
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

We can see that the VIF value for the above model is insignificant for the column "hum" so we will drop it for the best fit.

```python
X_train_new=X_train_new.drop("hum",axis=1)
```

## Model 6

```python
X_train_sm=sm.add_constant(X_train_new)
lr=sm.OLS(y_train,X_train_sm)
lr_model_6=lr.fit()
lr_model_6.summary()
```

```python
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

As we can observed that the p-value is significant for the variables and the VIF for the "temp" is greater than 5 but as we also know that the "temp" variable is crucial variable for the analysis so we will stop droping our variables further.

## <i> Residual Analysis</i>

```python
y_train_pred=lr_model_6.predict(X_train_sm)

res=y_train-y_train_pred

res
```

```python
sns.distplot(res)
```

From the above distplot we can conclude that the error terms are normally distributed at 0 mean which proves one of the assumption of the Linear regression.

```python
sns.scatterplot(y_train,res)
```

Conclusion: The error term are independent of each other which valids the assumption of linear regression.

```python
np.var(res)
```
